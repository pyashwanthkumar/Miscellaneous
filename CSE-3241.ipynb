### Accessing data stored in SQLite using Python and Pandas

```python
import pandas as pd
import sqlite3

# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("project.sqlite")
df = pd.read_sql_query("SELECT * from SALES", con)

# verify that result of SQL query is stored in the dataframe
#print(df.head())

con.close()
```

* First few rows of the DataFrame.

```python
df.head()
```

```python
data_sum = df.groupby("STORE_ID")[["Price"]].sum()
data_sum.head()
```

### Heatmap
* Below we have two heat maps, one grouped by the Price and one grouped by Quanity of Books sold.

```python
sns.heatmap(data_sum, cmap="Blues",xticklabels=False, cbar_kws={'label': 'Total Revenue Generated By Each Store'})
plt.ylabel("Store Number")
plt.title("Heat Map Showing Total Revenue Generated by Each Store")
plt.show()
```

```python
data_qty = df.groupby("STORE_ID")[["Quantity"]].sum()
data_qty.head()
```

```python
sns.heatmap(data_qty, cmap="Greens",xticklabels=False, cbar_kws={'label': 'Total Number of Books Sold'})
plt.ylabel("Store Number")
plt.title("Heat Map Showing Total No. of Books Sold in Each Store")
plt.show()
```
